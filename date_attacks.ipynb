```python
# Filter out provoked and boat attacks
data = data[(data['Type'] != 'Provoked') & (data['Type'] != 'Boat')]
# Extract the month from the "Date" column
data['Month'] = pd.to_datetime(data['Date'], format='%Y-%m-%d', errors='coerce').dt.month

# Exclude invalid dates
valid_data = data.dropna(subset=['Month'])

# Count the occurrences of each month
monthly_counts = valid_data['Month'].value_counts().sort_index()

# Print the monthly counts
print(monthly_counts)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('global-shark-attack.csv', delimiter=';')

# Filter out provoked and boat attacks
filtered_data = data[(data['Type'] != 'Provoked') & (data['Type'] != 'Boat')]

# Extract the month from the "Date" column
filtered_data['Month'] = pd.to_datetime(filtered_data['Date'], format='%Y-%m-%d', errors='coerce').dt.month

# Exclude invalid dates
valid_data = filtered_data.dropna(subset=['Month'])

# Count the occurrences of each month
monthly_counts = valid_data['Month'].value_counts().sort_index()

# Create a bar plot
plt.bar(monthly_counts.index, monthly_counts.values)

# Set the x-axis labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(monthly_counts.index, months)

# Set the y-axis label
plt.ylabel('Shark Attack Count')

# Set the title
plt.title('Monthly Shark Attack Counts (Excluding Provoked and Boat Attacks)')

# Display the plot
plt.show()
```
