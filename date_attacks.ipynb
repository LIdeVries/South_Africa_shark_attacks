```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('global-shark-attack.csv', delimiter=';')

# Filter out provoked and boat attacks
filtered_data = data[(data['Type'] != 'Provoked') & (data['Type'] != 'Boat')]

# Extract the month from the "Date" column
filtered_data['Month'] = pd.to_datetime(filtered_data['Date'], format='%Y-%m-%d', errors='coerce').dt.month

# Exclude invalid dates
valid_data = filtered_data.dropna(subset=['Month'])

# Count the occurrences of each month
monthly_counts = valid_data['Month'].value_counts().sort_index()

# Create a bar plot
plt.bar(monthly_counts.index, monthly_counts.values)

# Set the x-axis labels
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(monthly_counts.index, months)

# Set the y-axis label
plt.ylabel('Shark Attack Count')

# Set the title
plt.title('Monthly Shark Attack Counts (Excluding Provoked and Boat Attacks)')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('global-shark-attack.csv', delimiter=';')

# Group the data by 'Location' and count the occurrences
attacks_by_place = data['Location'].value_counts()

# Create a table of shark attacks by place
attacks_table = pd.DataFrame(attacks_by_place)
attacks_table.reset_index(inplace=True)
attacks_table.columns = ['Location', 'Attacks']

# Print the head of the table
print(attacks_table.head())
print(attacks_table.tail())
print(attacks_table)

# Sort the data for plotting
attacks_by_place_sorted = attacks_by_place.sort_values(ascending=False)

# Set the figure size
plt.figure(figsize=(100, 60))

# Create a bar plot for shark attacks by place
plt.bar(attacks_by_place_sorted.index, attacks_by_place_sorted.values)

# Set the x-axis label rotation
plt.xticks(rotation='vertical')

# Set the axis labels and title
plt.xlabel('Location')
plt.ylabel('Number of Shark Attacks')
plt.title('Shark Attacks by Place')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file
data = pd.read_csv('global-shark-attack.csv', delimiter=';')

# Group the data by 'Location' and count the occurrences
attacks_by_place = data['Location'].value_counts()

# Create a table of shark attacks by place
attacks_table = pd.DataFrame(attacks_by_place)
attacks_table.reset_index(inplace=True)
attacks_table.columns = ['Location', 'Attacks']

# Print the head of the table
print(attacks_table.head())

# Sort the data for plotting
attacks_by_place_sorted = attacks_by_place.sort_values(ascending=False)

# Set the figure size and bottom margin
plt.figure(figsize=(12, 6))
plt.subplots_adjust(bottom=0.3)

# Create a bar plot for shark attacks by place
plt.bar(attacks_by_place_sorted.index, attacks_by_place_sorted.values)

# Rotate and align the x-axis labels
plt.xticks(rotation=90, ha='center')

# Set the axis labels and title
plt.xlabel('Location')
plt.ylabel('Number of Shark Attacks')
plt.title('Shark Attacks by Place')

# Display the plot
plt.show()
```
